```python
# Inheritence
# Animal -> Dog -> Cat -> Mouse

class Animal : 
    def __init__(self,name):
        self.name=name
        self.favouriteFood="grass"
        
    def introduce(self):
        print("Hello I am "+self.name)
        
    def makeNoises(self):
        print("....")
        
class Dog : 
    def __init__(self,name):
        self.name=name
        self.favouriteFood="bone"
        self.collar="DOG"
    def introduce(self):
        print("Hello I am "+self.name)
    def makeNoises():
        print("Bark")       
        
class Cat : 
    def __init__(self,name):
        self.name=name
        self.favouriteFood="fish"
        self.litter="LITTERBOX"
    def introduce(self):
        print("Hello I am "+self.name)
        
    def makeNoises(self):
        print("Meow")     

```

```python
class Animal : # PARENT CLASS
    def __init__(self,name,favouriteFood):
        self.name=name
        self.favouriteFood=favouriteFood
        
    def introduce(self):
        print("Hello I am "+self.name)
        
    def makeNoises(self):
        print("....")


class Dog(Animal):
    def __init__(self,name,favouriteFood,collar):
        self.collar=collar
        Animal.__init__(self,name,favouriteFood)
        
    def makeNoises(self):
        print("BARK BARK")
        
class Cat(Animal):
    def __init__(self,name,favouriteFood,litter):
        self.litter=litter
        Animal.__init__(self,name,favouriteFood)
        
    def makeNoises(self):
        print("MEOW MEOW")
        

  
parrot=Animal("Paru" ,"Leaves")
parrot.introduce()
parrot.makeNoises() 


doggy=Dog("Rufus","Bone","1123")   
doggy.introduce()
doggy.makeNoises()  
        
catto=Cat("Mosfet","Chicken","12")   
catto.introduce()
catto.makeNoises()  
        
    
    
        
    
```

```python
## Multiple Inheritence
## Polymorphism
class Mammal:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def birth(self):
        print("Hello I'm a mammal and I was born")
    def introduce(self):
        print("Hello I'm a mammal")
        
class FlyingAnimal:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def fly(self):
        print("Hello I'm flying")
        
    def introduce(self):
        print("Hello I'm a flying animal")
        
class Bat(FlyingAnimal,Mammal):
    def __init__(self,a,b,c,x,y):
        Mammal.__init__(self,a,b,c)
        FlyingAnimal.__init__(self,x,y)


batman=Bat("a","b",2,3,4)
batman.birth()
batman.fly()
batman.introduce()

bird=FlyingAnimal(1,2)
#bird.fly()
```
